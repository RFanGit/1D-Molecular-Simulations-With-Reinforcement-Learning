```python
import numpy as np
import matplotlib.pyplot as plt

target_avgs = np.array([1, 2, 5, 10, 20, 30, 50])
reward_means_opt = np.loadtxt("reward_means_opt.txt")
reward_stds_opt = np.loadtxt("reward_std_opt.txt")

reward_means_lagg1 = np.loadtxt("reward_means_lagg1.txt")
reward_stds_lagg1 = np.loadtxt("reward_std_lagg1.txt")

reward_means_lagg1history1 = np.loadtxt("reward_means_lagg1history1.txt")
reward_stds_lagg1history1 = np.loadtxt("reward_std_lagg1history1.txt")

reward_means_lagg1history2 = np.loadtxt("reward_means_lagg1history2.txt")
reward_stds_lagg1history2 = np.loadtxt("reward_std_lagg1history2.txt")

reward_means_lagg1history2new = np.loadtxt("reward_means_lagg1history22.txt")
reward_stds_lagg1history2new = np.loadtxt("reward_std_lagg1history22.txt")

reward_means_lagg1history12 = np.loadtxt("reward_means_lagg1history12.txt")
reward_stds_lagg1history12 = np.loadtxt("reward_std_lagg1history12.txt")

reward_means_lagg2 = np.loadtxt("reward_means_lagg2.txt")
reward_stds_lagg2 = np.loadtxt("reward_std_lagg2.txt")

reward_means_lagg2history1 = np.loadtxt("reward_means_lagg2history1.txt")
reward_stds_lagg2history1 = np.loadtxt("reward_std_lagg2history1.txt")

reward_means_lagg2history2 = np.loadtxt("reward_means_lagg2history2.txt")
reward_stds_lagg2history2 = np.loadtxt("reward_std_lagg2history2.txt")

reward_means_lagg2history12 = np.loadtxt("reward_means_lagg2history12.txt")
reward_stds_lagg2history12 = np.loadtxt("reward_std_lagg2history12.txt")

# Plotting the means with error bars
plt.figure(figsize=(8, 6))
plt.errorbar(target_avgs, reward_means_opt, yerr=reward_stds_opt, fmt='o-',color='k', capsize=5, label='Optimal Solution')
plt.errorbar(target_avgs, reward_means_lagg1, yerr=reward_stds_lagg1, fmt='o--',color='b', capsize=5, label='Lagg 1')
plt.errorbar(target_avgs, reward_means_lagg1history1, yerr=reward_stds_lagg1history1, fmt='o--', capsize=5, label='Lagg 1 History 1')
plt.errorbar(target_avgs, reward_means_lagg1history2, yerr=reward_stds_lagg1history2, fmt='o--', capsize=5, label='Lagg 1 History 2')
plt.errorbar(target_avgs, reward_means_lagg1history12, yerr=reward_stds_lagg1history12, fmt='o--', capsize=5, label='Lagg 1 History 1 2')

#plt.errorbar(target_avgs, reward_means_lagg2, yerr=reward_stds_lagg2, fmt='o--', capsize=5, label='Lagg 2')
#plt.errorbar(target_avgs, reward_means_lagg2history1, yerr=reward_stds_lagg2history1, fmt='o--', capsize=5, label='Lagg 2 History 1')
#plt.errorbar(target_avgs, reward_means_lagg2history12, yerr=reward_stds_lagg2history2, fmt='o--', capsize=5, label='Lagg 2 History 2')
#plt.errorbar(target_avgs, reward_means_lagg2history2, yerr=reward_stds_lagg2history12, fmt='o--', capsize=5, label='Lagg 2 History 1 2')


# Adding labels and title
plt.xlabel('Target Average', fontsize = 20)
plt.ylabel('Variability', fontsize = 20)
plt.legend(loc='best', fontsize = 15)
plt.title("dt = 0.5", fontsize = 20)

# Show plot
plt.grid(True)
plt.savefig("dt_05_comparison_lagg1.png")
plt.show()
```
