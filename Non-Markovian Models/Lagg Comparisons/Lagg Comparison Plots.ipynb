```python
import numpy as np
import matplotlib.pyplot as plt

## Importing all our data

target_avgs = np.array([1, 2, 5, 10, 20, 30, 50])
reward_means_opt = np.loadtxt("reward_means_opt.txt")
reward_stds_opt = np.loadtxt("reward_std_opt.txt")

reward_means_lagg1 = np.loadtxt("reward_means_lagg1.txt")
reward_stds_lagg1 = np.loadtxt("reward_std_lagg1.txt")

reward_means_lagg1history1 = np.loadtxt("reward_means_lagg1history1.txt")
reward_stds_lagg1history1 = np.loadtxt("reward_std_lagg1history1.txt")

reward_means_lagg1history2 = np.loadtxt("reward_means_lagg1history2.txt")
reward_stds_lagg1history2 = np.loadtxt("reward_std_lagg1history2.txt")

reward_means_lagg1history12 = np.loadtxt("reward_means_lagg1history12.txt")
reward_stds_lagg1history12 = np.loadtxt("reward_std_lagg1history12.txt")

reward_means_lagg2 = np.loadtxt("reward_means_lagg2.txt")
reward_stds_lagg2 = np.loadtxt("reward_std_lagg2.txt")

reward_means_lagg2history1 = np.loadtxt("reward_means_lagg2history1.txt")
reward_stds_lagg2history1 = np.loadtxt("reward_std_lagg2history1.txt")

reward_means_lagg2history2 = np.loadtxt("reward_means_lagg2history2.txt")
reward_stds_lagg2history2 = np.loadtxt("reward_std_lagg2history2.txt")

reward_means_lagg2history12 = np.loadtxt("reward_means_lagg2history12.txt")
reward_stds_lagg2history12 = np.loadtxt("reward_std_lagg2history12.txt")

print ("Done!")
```

```python
# Plotting the means with error bars
plt.figure(figsize=(8, 6))
plt.errorbar(target_avgs, reward_means_opt, yerr=reward_stds_opt, fmt='o-',color='k', capsize=5, label='Optimal Controller')
plt.errorbar(target_avgs, reward_means_lagg1, yerr=reward_stds_lagg1, fmt='o--',color='b', capsize=5, label='Machine Learning (Lagg 1)')
plt.errorbar(target_avgs, reward_means_lagg2, yerr=reward_stds_lagg1, fmt='o--',color='r', capsize=5, label='Machine Learning (Lagg 1)')

# Adding labels and title
plt.xlabel('Target Average', fontsize = 20)
plt.ylabel('Variability', fontsize = 20)
plt.legend(loc='best', fontsize = 15)
plt.title("dt = 0.5", fontsize = 20)

# Show plot
plt.grid(True)
plt.savefig("Lagg_Makes_Worse_Controllers_dt05.png")
plt.show()
```

```python
# Plotting the means with error bars
avg_index = 2

plt.figure(figsize=(8, 6))

plt.errorbar(0, reward_means_opt[avg_index], yerr=reward_stds_opt[avg_index], fmt='o-',color='k', capsize=5, label='Optimal Controller')
plt.errorbar(1, reward_means_lagg1[avg_index], yerr=reward_stds_lagg1[avg_index], fmt='o--',color='b', capsize=5, label='Machine Learning ')
plt.errorbar(1, reward_means_lagg1history1[avg_index], yerr=reward_stds_lagg1history1[avg_index], color='r',fmt='o--', capsize=5, label='Machine Learning w/ History (1 step)')
plt.errorbar(1, reward_means_lagg1history2[avg_index], yerr=reward_stds_lagg1history2[avg_index], color='g',fmt='o--', capsize=5, label='Machine Learning w/ History (2 steps)')
plt.errorbar(1, reward_means_lagg1history12[avg_index], yerr=reward_stds_lagg1history12[avg_index], color='m',fmt='o--', capsize=5, label='Machine Learning w/ History (1+2 steps)')

plt.errorbar(2, reward_means_lagg2[avg_index], yerr=reward_stds_lagg2[avg_index], fmt='o--',color='b', capsize=5)#, label='Lagg 2')
plt.errorbar(2, reward_means_lagg2history1[avg_index], yerr=reward_stds_lagg2history1[avg_index],color='r', fmt='o--', capsize=5)#, label='History 1')
plt.errorbar(2, reward_means_lagg2history2[avg_index], yerr=reward_stds_lagg2history2[avg_index],color='g', fmt='o--', capsize=5)#, label='History 2')
plt.errorbar(2, reward_means_lagg2history12[avg_index], yerr=reward_stds_lagg2history12[avg_index],color='m', fmt='o--', capsize=5)#, label='History 1 2')

# Adding labels and title
#plt.ylim([0, reward_means_lagg2[avg_index]+0.5])


plt.xlabel('Controller Lagg', fontsize = 20)
plt.ylabel('Variability', fontsize = 20)
plt.legend(loc='best', fontsize = 15)
plt.title("Average = "+str(target_avgs[avg_index])+", dt = 0.5", fontsize = 20)

# Show plot
plt.grid(True)
plt.savefig("Comparison_Of_Avg5_dt05.png")
plt.show()
```

```python
baseline_lagg = reward_means_lagg1[avg_index]
minimized_lagg = min([reward_means_lagg1history1[avg_index], reward_means_lagg1history2[avg_index], reward_means_lagg1history12[avg_index]])
reduction = minimized_lagg/baseline_lagg

print ("The effect of a history dependent controller reduced the variability from ", baseline_lagg , " to ",minimized_lagg )
print ("The variability has decreased to ", reduction*100 ," percent of the original variability.")

##However the minimum lagg we could possibly get is given by reward_means_opt[avg_index]
##how can we track how close it gets to a particular number...?

##we can subtract both from the actual number and divide by the actual number for a percentage error. 
##Then we'll know how much percentage error there was
optimal = reward_means_opt[avg_index]
baseline_error = baseline_lagg/optimal - 1
history_error = minimized_lagg/optimal - 1
improvement_ratio = baseline_error/history_error - 1

print ("The baseline variability for this average and temporal resolution is ", optimal)
print ("The baseline error is ", baseline_error)
print ("The smart error is ", history_error)
print ("This is a ", improvement_ratio*100, " percent improvement using histories.")
```
